```python
import tensorflow as tf
```

```python
# TextVectorization hyper parameter
text_dataset = ["i am planning to learn data science", "do you know about sub word embeddings", "i have very little knowledge about tensorflow"]
max_vocab = 500
max_len = 5
```

```python
# Initiate layer
vectorize_layer = tf.keras.layers.experimental.preprocessing.TextVectorization(max_tokens=max_vocab,
                                                                               standardize="lower_and_strip_punctuation",
                                                                               split="whitespace",
                                                                               output_mode="int",
                                                                               output_sequence_length=max_len)
# Now to time consume these text_dataset to generate vocab and it's mapping
vectorize_layer.adapt(text_dataset)

# Vocabulary generated by layer
print(vectorize_layer.get_vocabulary())
```

```python
# To see output of the layer let's build small network
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(1,), dtype=tf.string))
model.add(vectorize_layer)

model.predict([["i have learn about the byte pair encoding"], ["data science is everywhere"]])
```
